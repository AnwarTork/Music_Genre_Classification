```python
from sklearn.pipeline import Pipeline
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder
from sklearn.metrics import accuracy_score,confusion_matrix
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
#from sklearn.metrics import plot_roc_curve,roc_auc_score,roc_curve
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder,StandardScaler,RobustScaler
from sklearn.compose import ColumnTransformer
```

```python
#music_data = music_data.drop(columns={'key','mode','instrumentalness','liveness'})
```

```python
data_label = music_data["Class"].copy()
music_data = music_data.drop("Class", axis=1)


categorical_pipeline = Pipeline(
    steps=[
        ("impute", SimpleImputer(strategy="most_frequent")),
        ("oh-encode", OneHotEncoder(handle_unknown="ignore", sparse=False)),
    ]
)


numeric_pipeline = Pipeline(
    steps=[("impute", SimpleImputer(strategy="mean")),
           ("scale", RobustScaler())]
)


cat_cols = music_data.select_dtypes(exclude="number").columns
num_cols = music_data.select_dtypes(include="number").columns

preprocessor = ColumnTransformer(
    transformers=[
        ("numeric", numeric_pipeline, num_cols),
        ("categorical", categorical_pipeline, cat_cols),
    ]
)

music_data_prepared = preprocessor.fit_transform(music_data)
music_data_test_prepared = preprocessor.transform(music_data_test)
```

```python
X_train, X_valid, y_train, y_valid = train_test_split(music_data_prepared, data_label, test_size=0.2, random_state=42)
```

## LogisticRegression vs OneVsRestClassifier

```python
from sklearn.linear_model import LogisticRegression
from sklearn.multiclass import OneVsRestClassifier
ovr_clf = OneVsRestClassifier(LogisticRegression(max_iter=1000, random_state=1))
ovr_clf.fit(X_train, y_train)
y_valid_pred = ovr_clf.predict(X_valid)
```

```python
from sklearn.metrics import accuracy_score
confusion_matrix(y_valid, y_valid_pred)
print(accuracy_score(y_valid,y_valid_pred))
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_valid,y_valid_pred))
```

```python
y_test_pred = ovr_clf.predict(music_data_test_prepared)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV


# Create a Logistic Regression model
model = LogisticRegression()

# Define hyperparameters to tune
param_grid = {'C': [0.001, 0.01, 0.1, 1, 10, 100], 'penalty': ['l1', 'l2']}

# Tune hyperparameters using GridSearchCV
grid_search = GridSearchCV(model, param_grid=param_grid, cv=5)
grid_search.fit(X_train, y_train)
y_pred_GS_LR = grid_search.predict(X_valid)
```

```python
auc_score_kn_GS_LR = accuracy_score(y_valid,y_pred_GS_LR)
```

```python
y_test_pred_GS_LR = grid_search.predict(music_data_test_prepared)
```

```python
# Print best hyperparameters and score
print("Best hyperparameters: ", grid_search.best_params_)
print("Best score: ", grid_search.best_score_)
```
